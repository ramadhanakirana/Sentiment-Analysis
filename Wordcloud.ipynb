### **Wordcloud Positif**

```python
from nltk import ngrams
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter #untuk hitung frekuensi
```

```python
# Menggabungkan semua token menjadi satu string
text_pos = ' '.join(' '.join(map(str, ulasan)) for ulasan in df_ulasan[df_ulasan['Kelas Label'] == 'positif']['Ulasan_Tokenization'])
```

```python
# Hapus kata yang tidak diinginkan
words_to_remove = ['terima', 'kasih', 'tidak', 'oke', 'bantu']
for word in words_to_remove:
    text_pos = text_pos.replace(word, '')
```

```python
from collections import Counter

# Buat n-gram
# Membuat 2-gram dari token
two_grams_pos = ngrams(text_pos.split(), 1)

# Menghitung frekuensi kemunculan setiap n-gram
ngram_freq = Counter([' '.join(ngram) for ngram in two_grams_pos])

# Cetak frekuensi kemunculan kata-kata
for word, frequency in ngram_freq.most_common():
    print(word, frequency)
```

```python
# Buat n-gram
# Membuat 2-gram dari token
two_grams_pos = ngrams(text_pos.split(), 2)

# Menghitung frekuensi kemunculan setiap n-gram
ngram_freq = Counter([' '.join(ngram) for ngram in two_grams_pos])

# Menghapus n-gram yang memiliki kata yang sama berulang
unique_ngram_freq = {ngram: freq for ngram, freq in ngram_freq.items() if len(set(ngram.split())) == len(ngram.split())}

# Membuat WordCloud
wordcloud_pos = WordCloud(width=800, height=500, background_color='white', max_words=150)

# Membuat WordCloud dari frekuensi n-gram
wordcloud_pos.generate_from_frequencies(frequencies=unique_ngram_freq)

# Menampilkan WordCloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.axis('off')
plt.show()
```
